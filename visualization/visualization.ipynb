```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
import numpy as np
import os
```

```python
#read the csv file using pandas
df=pd.read_csv('./AllSTPrecipitation.csv')
df.head()
```

```python
#changing the format of date using pandas .to_datetime
df['Time']=pd.to_datetime(df['Time'])
df.head()
```

```python
df1=df.set_index('Time')
df1.head()
```

```python
#resampling the monthly data and sum these all precipitation value to show the total monthly trend

monthly=df1.resample('M').sum()
monthly
```

```python
#exporting the monthly pprecipitaiton data 
# monthly.to_csv('./MothlyPrecipitationData.csv')
```

```python
plot=monthly.plot(figsize=(25,10), marker='+')
plt.xlabel('Date', fontsize=12)
plt.ylabel('Total Precipitation(in mm) Value at Different Stations',fontsize=12)
plt.title('Monthly Precipitation Trend from 2015 to 2022 Among Different Stations',fontsize=14, fontweight='bold')
plt.legend()


#to save the plot in hd
# plt.savefig('./images/sample.png',dpi=800)
```

# seasonal temperature and precipitation analysis
 

```python
mapping = {1 : "Winter",2:"Winter",12: "Winter",3:"Spring",4:"Spring",10:"Late autumn",
           11:"Late autumn",8:"Autumn",9:"Autumn",6:"Rainy",7:"Rainy",4:"Summer",5:"Summer"}
x["Season"] = x["Month"].map(mapping).astype(str)
```

```python
x.head()
```

```python
pio.templates.default = "simple_white"
fig = px.pie(data_frame=x,values="Sum of Precipitation ",names='Season',labels= {"Season"},
             title="Pie chart of Precipitation  according to  Season from 2015 to 2022 August")
fig.update_traces(textposition ='inside',textinfo='percent+label+value')
fig.show()
```

```python
pio.templates.default = "simple_white"
fig = px.pie(data_frame=x,values="Average Max. Temperature",names='Season',labels= {"Season"},
             title="Pie chart of Max. Temperature according to  Season from 2015 to 2022 August")
fig.update_traces(textposition ='inside',textinfo='percent+label+value')
fig.show()
```
